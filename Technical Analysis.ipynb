```python
#Bargraph of Mutual Fund Technical Analysis
```

```python
plt.bar([1,3,5],[7,9,11], label="Trend")
plt.bar([2,4,6],[8,10,12], label="returns")
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Chart of Technical Analysis')
plt.show()
```
